```python
#installing packages 

import numpy as np
import pandas as pd
from pandas_datareader import data as web
import pandas_datareader
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# set the dates for the analysis

start = datetime.datetime(2014, 1, 1)
end = datetime.datetime(2019, 1, 1)
```

```python
# select the companies

CocaCola = web.DataReader('K', 'yahoo', start, end)
Pepsi = web.DataReader('PEP', 'yahoo', start, end)
```

```python
CocaCola.tail()
```

```python
Pepsi.tail()
```

# Line Plot with Two or more companies

```python
CocaCola['Close'].plot(label='CocaCola',figsize=(16,8))
Pepsi['Close'].plot(label='Pepsi')

_ = plt.xlabel('Date')

_ = plt.ylabel('Closing Price')

_ = plt.title('Pepsi & Coca-Cola: Price Comparison')

plt.legend();
```

# Creating a Volume Plot with two or more companies

```python
CocaCola['Volume'].plot(label='CocaCola',figsize=(16,8))
Pepsi['Volume'].plot(label='Pepsi')

_ = plt.xlabel('Date')

_ = plt.ylabel('Volume')

_ = plt.title('Pepsi & Coca-Cola: Volume Comparison')

plt.legend();
```

# Total Traded  Plot with two or more companies

```python
CocaCola['Total Traded'] = CocaCola['Close']*CocaCola['Volume']
Pepsi['Total Traded'] = Pepsi['Close']*Pepsi['Volume']
```

```python
CocaCola['Total Traded'].plot(label='Coca Cola',figsize=(16,8))
Pepsi['Total Traded'].plot(label='Pepsi')

_ = plt.xlabel('Date')

_ = plt.ylabel('Total Traded')

_ = plt.title('Pepsi & Coca-Cola: Total Traded')

plt.legend();
```

# Box-Plot with two or more companies

```python
united_box = pd.concat([Pepsi['Close'],CocaCola['Close']],axis=1)
united_box.columns = ['Pepsi Closing Price','Coca Cola Closing Price']
united_box.plot(kind='box',figsize=(8,11),colormap='jet')

_ = plt.ylabel('Closing Prices')

_ = plt.title('Pepsi and Coca Cola - Boxplot')
```

# Histogram (percent change) with two or more companies

```python
Pepsi['Returns'] = Pepsi['Close'].pct_change(1)
CocaCola['Returns'] = CocaCola['Close'].pct_change(1)
```

```python
Pepsi.tail()
```

```python
CocaCola.tail()
```

```python
Pepsi['Returns'].hist(bins=50);

_ = plt.xlabel('Returns')

_ = plt.ylabel('Frequency')

_ = plt.title('Pepsi: Histogram')
```

```python
CocaCola['Returns'].hist(bins=50);

_ = plt.xlabel('Returns')

_ = plt.ylabel('Frequency')

_ = plt.title('Coca-Cola: Histogram')
```

```python
CocaCola['Returns'].hist(bins=100,label='Coca-Cola',alpha=0.5)
Pepsi['Returns'].hist(bins=100,label='Pepsi',figsize=(10,8),alpha=0.5)

_ = plt.xlabel('Stock Return')

_ = plt.ylabel('Frequency')

_ = plt.title('Coca-Cola & Pepsi - Histogram')

plt.legend();
```


# Histogram (logarithmic return) with two or more companies

```python
Pepsi['LN Returns'] = np.log(Pepsi['Close']/Pepsi['Close'].shift(1))
CocaCola['LN Returns'] = np.log(CocaCola['Close']/CocaCola['Close'].shift(1))
```

```python
Pepsi.tail()
```

```python
CocaCola.tail()
```

```python
CocaCola['LN Returns'].hist(bins=50);

_ = plt.xlabel('Ln Returns')

_ = plt.ylabel('Frequency')

_ = plt.title('Coca-Cola: Histogram')
```

```python
Pepsi['LN Returns'].hist(bins=50);

_ = plt.xlabel('Returns')

_ = plt.ylabel('LN Frequency')

_ = plt.title('Pepsi: Histogram')
```

```python
Pepsi['LN Returns'].hist(bins=100,label='Pepsi',figsize=(10,8),alpha=0.5)
CocaCola['LN Returns'].hist(bins=100,label='Coca-Cola',alpha=0.5)
plt.legend();
```

```python
CocaCola['LN Returns'].hist(bins=100,label='Coca-Cola',alpha=0.5)
Pepsi['LN Returns'].hist(bins=100,label='Pepsi',figsize=(10,8),alpha=0.5)

_ = plt.xlabel('LN Stock Return')

_ = plt.ylabel('Frequency')

_ = plt.title('Coca-Cola & Pepsi - Histogram')

plt.legend();
```

# Kernel Density Plot (KDE)

```python
CocaCola['Returns'].plot(kind='kde',label='Coca-Cola',figsize=(12,6))
Pepsi['Returns'].plot(kind='kde',label='Pepsi')



_ = plt.xlabel('Stock Return')

_ = plt.ylabel('Density')

_ = plt.title('Coca-Cola & Pepsi - Histogram')

plt.legend();
```
