```python
#installing packages 

import numpy as np
import pandas as pd
from pandas_datareader import data as web
import pandas_datareader
import datetime
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm
%matplotlib inline
```

```python
start = datetime.datetime(2018, 1, 2)
end = datetime.datetime(2019, 4, 1)

tickers = ['AAPL','WMT', 'TM','KO','BA']
stocks = pd.DataFrame()
for x in tickers:
    stocks[x] = web.DataReader(x, 'yahoo', start, end)['Close']

stocks.tail()
```

```python
stocks_return = (stocks/stocks.shift(1))-1
stocks_return.tail()
```

```python
portfolio_weights = np.array(np.random.random(5))
portfolio_weights
portfolio_weights = portfolio_weights/np.sum(portfolio_weights)
portfolio_weights
```

```python
weighted_returns_portfolio = stocks_return.mul(portfolio_weights, axis = 1)
```

```python
#Calculating the returns of the portfolio

stocks_return['Portfolio'] = weighted_returns_portfolio.sum(axis=1).dropna()
```

```python
mu = stocks_return['Portfolio'].mean()
mu
```

```python
sigma = stocks_return['Portfolio'].std()
sigma
```

```python
confidence = 0.95
confidence
```

```python
position = 1e6
position
```

```python
T = 252
T
```

```python
S0 = stocks.mean().mean()
S0

```

```python
random_returns = np.random.normal(mu,sigma,T) + 1
```

```python
forecasted_values = random_returns.cumprod() * S0
```

```python
plt.plot(range(0, T), forecasted_values)
plt.show()
```

```python
#creating a montecarlo simulation

# Loop through 100 simulations
for i in range(100):

    # Generate the random returns
    random_returns = np.random.normal(mu, sigma, T) + 1
    
    # Create the Monte carlo path
    forecasted_values = S0*(random_returns).cumprod()
    
    # Plot the Monte Carlo path
    plt.plot(range(T), forecasted_values)

# Show the simulations
plt.show()
```

```python
# Aggregate the returns
sim_returns = []

# Loop through 100 simulations
for i in range(100):

    # Generate the Random Walk
    random_returns = np.random.normal(mu, sigma, T)
    
    # Save the results
    sim_returns.append(random_returns)

# Calculate the VaR(99)
var_99 = np.percentile(sim_returns, 1)
print("Parametric VaR(99): ", round(100*var_99, 2),"%")
```
