# Creating a Line Plot and calculating range

```python

# libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## getting data from Tesla

import pandas_datareader
import datetime

import pandas_datareader.data as web
```

```python
start = datetime.datetime(2015,1,1)
end = datetime.datetime(2019,1,1)

Tesla = web.DataReader('TSLA','yahoo',start,end)
```

```python
#with a dummy variable

_ = plt.plot(Tesla['Close'])

_ = plt.xlabel('Date')

_ = plt.ylabel('Closing Price')

_ = plt.title('Tesla Closing Price')

```

```python
#more descriptive linear plot

Tesla['Close'].plot(label='Tesla',figsize=(19,8),title='Tesla Closing Price')

_ = plt.xlabel('Date')

_ = plt.ylabel('Closing Price')
```
