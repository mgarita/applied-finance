```python
# libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# getting data from Tesla

import pandas_datareader
import datetime

import pandas_datareader.data as web
```

```python
start = datetime.datetime(2015,1,1)
end = datetime.datetime(2019,1,1)

Tesla = web.DataReader('TSLA','yahoo',start,end)
```

```python
Tesla.head()
```

```python
## Plotting the volume 

Tesla['Volume'].plot(label='Tesla',figsize=(16,8),title='Volume Traded')

_ = plt.xlabel('Date')

_ = plt.ylabel('Volume')
```

```python

Tesla['Close'].plot(label='Tesla',figsize=(16,8),title='Closing Price')

_ = plt.xlabel('Date')

_ = plt.ylabel('Price')
```

```python
Tesla['Volume'].argmax()
```

```python
Tesla['Volume'].argmin()
```

# Total Money Traded

```python
Tesla_total_traded = Tesla['Close'] * Tesla['Volume']
```

```python
Tesla_total_traded.head()
```

```python
Tesla_total_traded.plot(label='Tesla',figsize=(16,8),title='Total Traded')

_ = plt.xlabel('Date')

_ = plt.ylabel('Total Traded')
```
